```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('covid-chestxray-dataset/metadata.csv')
df.columns
```

```python
sns.pairplot(df, hue='finding')
```

```python
plt.figure(figsize = (16,5))
sns.heatmap(
  df.corr(),
  annot=True, 
  linewidths=.5, 
  cmap=sns.diverging_palette(220, 20, as_cmap=True),
  center=0
)
```
