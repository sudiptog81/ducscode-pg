Breast Cancer Dataset

```python
from scipy.cluster import hierarchy
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

```python
z = hierarchy.linkage(df.data, 'average')
z
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
dn = hierarchy.dendrogram(z, orientation='right')
```

Seeds Dataset

```python
from pycaret.datasets import get_data

get_data('index')
```

```python
df_seeds = get_data('seeds')
```

```python
import numpy as np
import seaborn as sns

sns.heatmap(np.isfinite(df_seeds))
```

```python
df_seeds = df_seeds.dropna()
```

```python
df_seeds.plot.scatter(x=2, y=6)
```

```python
z_seeds = hierarchy.linkage(df_seeds.to_numpy(), 'complete')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
dn_seeds = hierarchy.dendrogram(z_seeds, orientation='right')
```
